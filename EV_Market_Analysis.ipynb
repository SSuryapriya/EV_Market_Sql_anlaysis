```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/global_ev_market_charging_infrastructure_2026.csv")
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
print(df.describe())
```

```python
top_brands = df.groupby('ev_brand')['ev_sales_units'].sum().sort_values(ascending=False).head(10)
print(top_brands)
```

```python
country_sales = df.groupby('country')['ev_sales_units'].sum().sort_values(ascending=False)
print(country_sales)
```

```python
region_sales = df.groupby('region')['ev_sales_units'].sum().sort_values(ascending=False)
print(region_sales)
```

```python
print(df['battery_capacity_kwh'].mean())
```

```python

print(df['charging_time_hours'].mean())
```

```python
print(df['charging_stations'].sum())
```

```python
top_brands.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 EV Brands by Sales")
plt.xlabel("EV Brand")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
country_sales.head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Countries by EV Sales")
plt.xlabel("Country")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['vehicle_type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Vehicle Type Distribution")
plt.ylabel("")
plt.show()
```
